```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
df = pd.DataFrame(purchase_data)
df.head()
```

## Player Count

* Display the total number of players


```python
players = df["SN"].nunique()
totalplayers = {"Total Players": [players]}
tp_df = pd.DataFrame(totalplayers) 
tp_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
uniqueitems = df["Item Name"].nunique()
avgprice = df["Price"].mean()
purchases = df["Purchase ID"].sum()
revenue = df["Price"].sum()

purchasinganalysis = {"Number of Unique Items": [uniqueitems],
                       "Average Price": [avgprice],
                       "Number of Purchases": [purchases],
                       "Total Revenue": [revenue]}

patotal_df = pd.DataFrame(purchasinganalysis)
patotal_df["Average Price"] = patotal_df["Average Price"].astype(float).map("${:,.2f}".format)
patotal_df["Total Revenue"] = patotal_df["Total Revenue"].astype(float).map("${:,.2f}".format)

patotal_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# df["Gender"].value_counts()
# Male                     652
# Female                   113
# Other / Non-Disclosed     15
male_df = df.loc[df["Gender"] == "Male", :]
count_male = male_df["SN"].nunique()
percentmale = (count_male / players) *100

female_df = df.loc[df["Gender"] == "Female", :]
count_female = female_df["SN"].nunique()
percentfemale = (count_female / players) *100


othergender_df = df.loc[df["Gender"] == "Other / Non-Disclosed", :]
count_og = othergender_df["SN"].nunique()
percentother = (count_og / players) *100

genderdemos = {"Gender": ["Male","Female","Other / Non-Disclosed"],
               "Total Count": [count_male, count_female, count_og],
               "Percentage of Players": [percentmale, percentfemale, percentother]
}

genderdemo_df = pd.DataFrame(genderdemos)
genderdemo_df["Percentage of Players"] = genderdemo_df["Percentage of Players"].astype(float).map("{:,.2f}".format)
genderdemoindex_df= genderdemo_df.set_index("Gender")
genderdemoindex_df
```

## Age Demographics

```python
# * Establish bins for ages
bins = [0,9,14,19,24,29,34,39,100]
bin_names = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
# * Categorize the existing players using the age bins. Hint: use pd.cut()
df["Age Group"] = pd.cut(df["Age"], bins, labels=bin_names)
df.head()
unique_users = df.drop_duplicates(["SN"])

# * Calculate the numbers and percentages by age group
age_group = unique_users.groupby(["Age Group"])
agegroupcount = age_group.count()
age_counts = [agegroupcount.loc[x, "SN"] for x in bin_names]
age_percents = [100*x/players for x in age_counts]

# * Create a summary data frame to hold the results

age_demos = {"Age Group": bin_names,
                 "Total Count": age_counts,
                 "Percentage of Players": age_percents
            }
agedemos_df = pd.DataFrame(age_demos)
agedemos_df= agedemos_df.set_index("Age Group")
# * Optional: round the percentage column to two decimal points
agedemos_df["Percentage of Players"] = agedemos_df["Percentage of Players"].astype(float).map("{:,.2f}".format)

# * Display Age Demographics Table
agedemos_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Bin the purchase_data data frame by age

agebins = df.groupby(["Age Group"])



# Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below
agecountsummary = agebins.count()
age_purch_count = [agecountsummary.loc[x, "Purchase ID"] for x in bin_names]


agesumsummary = agebins.sum()
age_total_purch = [agesumsummary.loc[x, "Price"] for x in bin_names]

agemeansummary = agebins.mean()
age_avg_purch = [agemeansummary.loc[x, "Price"] for x in bin_names]

age_avg_per_person = [agesumsummary.loc[x,"Price"]/agedemos_df.loc[x,"Total Count"] for x in bin_names]

# Create a summary data frame to hold the results
agepurchanalysis = {"Age Group": bin_names,
                 "Purchase Count": age_purch_count,
                 "Average Purchase Price": age_avg_purch,
                 "Total Purchase Value":age_total_purch,
                 "Avg Total Purchase per Person": age_avg_per_person}

agepurchanalysis_df = pd.DataFrame(agepurchanalysis)
# * Optional: give the displayed data cleaner formatting
agepurchanalysis_df["Average Purchase Price"] = agepurchanalysis_df["Average Purchase Price"].astype(float).map("${:,.2f}".format)
agepurchanalysis_df["Total Purchase Value"] = agepurchanalysis_df["Total Purchase Value"].astype(float).map("${:,.2f}".format)
agepurchanalysis_df["Avg Total Purchase per Person"] = agepurchanalysis_df["Avg Total Purchase per Person"].astype(float).map("${:,.2f}".format)

agepurchanalysissorted_df = agepurchanalysis_df.sort_values(["Age Group"], ascending=True)
paagesortedindex_df= agepurchanalysissorted_df.set_index("Age Group")
# * Display the summary data frame
paagesortedindex_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# * Run basic calculations to obtain the results in the table below
uniqueusers = df["SN"].unique()

topusers = df.groupby(["SN"])

ucounts = topusers.count()
usercounts = [ucounts.loc[x, "Price"] for x in uniqueusers]

umean = topusers.mean()
usermeans = [umean.loc[x, "Price"] for x in uniqueusers]

usum = topusers.sum()
usertotals = [usum.loc[x, "Price"] for x in uniqueusers]

# * Create a summary data frame to hold the results

topuseranalysis = {"SN": uniqueusers,
                 "Purchase Count": usercounts,
                 "Average Purchase Price": usermeans,
                 "Total Purchase Value": usertotals}

topuser_df = pd.DataFrame(topuseranalysis)

# * Sort the total purchase value column in descending order

topusersorted_df = topuser_df.sort_values(["Total Purchase Value"], ascending=False)

# * Optional: give the displayed data cleaner formatting
topusersorted_df["Average Purchase Price"] = topuser_df["Average Purchase Price"].astype(float).map("${:,.2f}".format)
topusersorted_df["Total Purchase Value"] = topuser_df["Total Purchase Value"].astype(float).map("${:,.2f}".format)
topusersorted_df= topusersorted_df.set_index("SN")

# * Display a preview of the summary data frame
topusersorted_df.head()
```
